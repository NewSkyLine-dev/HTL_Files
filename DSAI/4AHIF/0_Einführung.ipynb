# Lineare Regression pure Python

```python
import math

x = [1.9, 2.5, 3.2, 3.8, 4.7, 5.5, 5.9, 7.2]
y = [22, 33, 30, 42, 38, 49, 42, 55]

def mean(values):
 return sum(values) / float(len(values))

def covariance(x, mean_x, y, mean_y):
 covar = 0.0
 for i in range(len(x)):
  covar += (x[i] - mean_x) * (y[i] - mean_y)
 return covar

def variance(values, mean):
 return sum([(x-mean)**2 for x in values])

def linreg(x, y):
 x_mean, y_mean = mean(x), mean(y)
 b1 = covariance(x, x_mean, y, y_mean) / variance(x, x_mean)
 b0 = y_mean - b1 * x_mean
 return [b0, b1]

a, b = linreg(x, y)
print (f'b: {b} a: {a}')
```

# Daten für Frequency Beispiel auf numpy Basis

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.array([0, 5, 1, 10, 3, 7, 9, 2])
y = np.array([55, 10, 44, 14, 20, 50, 21, 4])
```

# Streudiagramme und lineare Regression mit matplotlib und numpy

```python
# Grundlegendes Streudiagramm erstellen
plt.xlabel('Sonnenstunden = unbhängige Variable')
plt.ylabel('Konzertbesucher (in 100) = abhängige Variable')
plt.plot(x, y, 'o')
```

```python
#Basisstreudiagramm erstellen
plt.plot(x, y, 'o')
# Erhalten Sie m (Steigung) und b (Achsenabschnitt) der linearen Regressionslinie
m, b = np.polyfit(x, y, 1)
# Lineare Regressionslinie zum Streudiagramm hinzufügen
plt.plot(x, m*x+b)
print (f'b: {m} a: {b}')
```

# Lineare Regression mit Scikit

```python
from sklearn.linear_model import LinearRegression

# x array muss 2 dimensional sein => reshape
xr = x.reshape((-1, 1))
print(xr)
print(y)
```

Typischer Ablauf:


1.   Trainiere Model mit alten Daten => fit
2.   Güte des Models => score
3.   Anwendung auf neue Daten => predict



```python
model = LinearRegression().fit(xr, y)
```

```python
r_sq = model.score(xr, y)
print(f"r2: {r_sq}")
corcoef = np.corrcoef(x, y)[0,1]
print(f"r: {corcoef}")
```

```python
x_pred = np.array([6]).reshape((-1,1))
y_pred = model.predict(x_pred)
print(f"predicted response:\n{y_pred}")
```
